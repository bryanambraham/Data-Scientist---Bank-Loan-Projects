SPLIT DATA TRAIN & TESTING

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

#Split data menjadi fitur (X) dan target (y)
X = df.drop('y', axis=1) #ini untuk ambil semua fitur kecuali "y"
y = df['y'] #Target

#Membagi dataset menjadi training dan testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, train_size=0.8, random_state=42)
```

LOGISTIC REGRESSION

```python
# Transform kategorikal menjadi numerikal untuk Model LogRegression

# One-Hot Encoding untuk kolom kategorikal
df = pd.get_dummies(df, columns=['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome'], drop_first=True)
df['y'] = df['y'].apply(lambda x: 1 if x == 'yes' else 0)
```

```python
from sklearn.linear_model import LogisticRegression

#Inisialisasi dan latih model Logistic Regression
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)

#Prediksi hasil
y_pred = logreg.predict(X_test)

#Evaluation Model
print('Logistic Regression Classification Report:\n', classification_report(y_test, y_pred))
print('Confusion Matrix: \n', confusion_matrix(y_test, y_pred))
```

1. Classification Report memberikan metrik precision, recall, dan f1-score untuk setiap kelas (di sini, 0 dan 1), serta average untuk kedua kelas.

Interpretasi untuk Kelas 0 (Negatif):
Precision (0.91): Dari semua prediksi yang dibuat oleh model untuk kelas 0 (negatif), 91% adalah benar (artinya, model memiliki tingkat ketepatan yang tinggi dalam memprediksi kelas 0).

Recall (0.97): Dari semua kasus yang sebenarnya kelas 0, model berhasil 97% di antaranya dengan benar. Artinya, model sangat baik dalam menangkap semua kasus negatif.

F1-Score (0.94): Ini adalah rata-rata harmonis antara precision dan recall. F1-score ini menunjukkan bahwa model sangat baik dalam memprediksi kelas 0, dengan nilai yang sangat seimbang antara precision dan recall.
